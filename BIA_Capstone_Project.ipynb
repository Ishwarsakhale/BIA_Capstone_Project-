# Treating missing values

```python
car_df.isna().sum()
```

the notRepairedDamage colum is easy to impute with mode 

```python
print(car_df.shape)
car_df.notRepairedDamage = car_df['notRepairedDamage'].fillna(method='ffill')
car_df.isna().sum()
```

We have 4 variables left now with missing values - vehicleType, model, gearbox, fuelType

To impute the categorical variable we use decision tree classifier 

```python
from sklearn.tree import DecisionTreeClassifier, DecisionTreeRegressor
from sklearn.model_selection import train_test_split
```

```python
# vehicleType 

# filter data for model building
X = car_df[car_df['vehicleType'].notna()].drop(['vehicleType','model'], axis=1)
y = car_df.vehicleType[car_df['vehicleType'].notna()]

# create test data with unknown vehicleType fields
xt = car_df[car_df['vehicleType'].isna()].drop(['vehicleType','model'], axis=1)

# one-hot encode the categorical variables
X_dum = pd.get_dummies(X)
xt_dum = pd.get_dummies(xt)

# build a decision tree model
dtree_veh = DecisionTreeClassifier().fit(X_dum,y)

# predict on test data
veh_pred = dtree_veh.predict(xt_dum)

# fill the missing values with the predicted values
car_df.vehicleType[car_df['vehicleType'].isna()] = veh_pred
car_df.vehicleType.isna().sum()
```

```python
# model 

# filter data for model building
X = car_df[car_df['model'].notna()].drop(['model'], axis=1)
y = car_df.model[car_df['model'].notna()]

# create test data with unknown model fields
xt = car_df[car_df['model'].isna()].drop(['model'], axis=1)

# one-hot encode the categorical variables
X_dum = pd.get_dummies(X)
xt_dum = pd.get_dummies(xt)
xt_dum.drop('brand_sonstige_autos', axis=1, inplace=True)

# build a decision tree model
dtree_model = DecisionTreeClassifier().fit(X_dum,y)

# predict on test data
model_pred = dtree_model.predict(xt_dum)

# fill the missing values with the predicted values
car_df.model[car_df['model'].isna()] = model_pred
car_df.model.isna().sum()
```

```python
# gearbox 

# filter data for model building
X = car_df[car_df['gearbox'].notna()].drop(['gearbox'], axis=1)
y = car_df.gearbox[car_df['gearbox'].notna()]

# create test data with unknown gearbox fields
xt = car_df[car_df['gearbox'].isna()].drop(['gearbox'], axis=1)

# one-hot encode the categorical variables
X_dum = pd.get_dummies(X)
xt_dum = pd.get_dummies(xt)
for col in X_dum.columns:
    if col not in xt_dum:
        xt_dum[col] = np.zeros(len(xt_dum))
        
# build a decision tree model
dtree_gear = DecisionTreeClassifier().fit(X_dum,y)

# predict on test data
gear_pred = dtree_gear.predict(xt_dum)

# fill the missing values with the predicted values
car_df.gearbox[car_df['gearbox'].isna()] = gear_pred
car_df.gearbox.isna().sum()
```

```python
# fuelType 

# filter data for model building
X = car_df[car_df['fuelType'].notna()].drop(['fuelType'], axis=1)
y = car_df.fuelType[car_df['fuelType'].notna()]

# create test data with unknown fuelType fields
xt = car_df[car_df['fuelType'].isna()].drop(['fuelType'], axis=1)

# one-hot encode the categorical variables
X_dum = pd.get_dummies(X)
xt_dum = pd.get_dummies(xt)
for col in X_dum.columns:
    if col not in xt_dum:
        xt_dum[col] = np.zeros(len(xt_dum))
        
# build a decision tree model
dtree_fuel = DecisionTreeClassifier().fit(X_dum,y)

# predict on test data
fuel_pred = dtree_fuel.predict(xt_dum)

# fill the missing values with the predicted values
car_df.fuelType[car_df['fuelType'].isna()] = fuel_pred
car_df.fuelType.isna().sum()
```
